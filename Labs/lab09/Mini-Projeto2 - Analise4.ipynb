```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Pacote que irá permitir visualizar todas linhas/colunas do dataframe
from IPython.display import display

from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

# Calcule a média de preço por marca e por veículo


```python
#configura para visualizar todas as linhas do dataframe
#pd.options.display.max_rows=None

frame = pd.DataFrame({'price':df['price'],
        'brand':df['brand'],
        'type':df['vehicleType']
        })

#defino as chaves de agrupament
key=['brand','type']

#realiza agrupamento calculando a media e definindo a escala com 2 casas decimais
frame.groupby(key, observed=True)['price'].mean().round(2)

#trata a coluna 'price' 
frame['price'].fillna(0,inplace=True)
frame["price"].isnull().value_counts()

#cria um dataframe final com a média de preço por marca e veículo
frame2 = pd.DataFrame(frame.groupby(key)['price'].mean().round(2))
frame2
```
